```python
import numpy as np
import pandas as pd
```

```python
movies_df=pd.read_csv("movies.csv")
```

```python
movies_df.head()
```

```python
ratings_df=pd.read_csv("ratings.csv")
```

```python
ratings_df.head()
```

```python
movie_rating_df=pd.merge(ratings_df,movies_df,on='movieId')
movie_rating_df.head()
```

```python
columns=['timestamp','genres']
movie_rating_df=movie_rating_df.drop(columns,axis=1)
```

```python
movie_rating_df.head()
```

```python
movie_rating_df=movie_rating_df.dropna(subset=['title'],axis=0)
```

```python
movie_rating_count=(movie_rating_df.groupby(['title'])['rating']
                    .count()
                    .reset_index().
                    rename(columns={'rating' : 'total_rate_count'})
                    [['title','total_rate_count']])
```

```python
movie_rating_count.head()
```

```python
rating_movie_ratingCount_df=movie_rating_df.merge(movie_rating_count,how='left',left_on='title',right_on='title')
```

```python
rating_movie_ratingCount_df.head(20)
```

```python
rating_movie_ratingCount_df=rating_movie_ratingCount_df.drop_duplicates(['userId','title'])
```

```python
rating_movie_ratingCount_df.head(20)
```

```python
movie_user_rating_pivot=rating_movie_ratingCount_df.pivot(index='userId',columns='title',values='rating').fillna(0)
```

```python
movie_user_rating_pivot.head(20)
```

```python
movie_user_rating_pivot.shape
```

```python
movie_user_rating_pivot_transpose=movie_user_rating_pivot.values.T
movie_user_rating_pivot_transpose.shape
```

```python
movie_user_rating_pivot_transpose
```

```python
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=12, random_state=17)
matrix = SVD.fit_transform(movie_user_rating_pivot_transpose)
matrix.shape
```

```python
correlation_coeff=np.corrcoef(matrix)
correlation_coeff.shape
```

```python
movie_titles=movie_user_rating_pivot.columns
movie_title_list=list(movie_titles)
movie_title_list
```

```python
import random
movie_index=random.randint(0,len(movie_title_list)-1)
print(movie_index)
movie_title=movie_title_list[movie_index]
print(movie_title)
```

```python
recommended_movies_corr=correlation_coeff[movie_index]
recommended_movies=list(movie_titles[(recommended_movies_corr>=0.9)])
recommended_movies
```
